```python
import nltk
import torch
import torch.nn as nn
import urllib.request
import bs4
import string
import matplotlib.pyplot as plt
```

```python
scrapped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/French_Revolution')
article = scrapped_data.read()

parsed_article = bs4.BeautifulSoup(article,'lxml')
paragraphs = parsed_article.find_all("p")

articles = ""

for p in paragraphs:
    articles += p.text
    
train_data = articles.translate(str.maketrans('','',string.punctuation)).lower()
train_data = train_data.translate(str.maketrans('','',string.digits))

text = nltk.word_tokenize(train_data)
stop_words = set(nltk.corpus.stopwords.words('english'))

text = [words for words in text if words not in stop_words]

vocab = set(text)
vocab_size = len(vocab)
```

```python
window = 2
ls = []

for center_index in range(len(text)):
    
    for sub in range(max(0, center_index - window), center_index):
        ls.append((center_index, sub))
        
    for add in range(center_index + 1, min(vocab_size - 1, center_index + window)+1):
        ls.append((center_index, add))
```

```python
word2int = { word : i for i,word in enumerate(vocab)}
int2word = {i : word for i,word in enumerate(vocab)}
```

```python
class Model(nn.Module):
    def __init__(self, vocab_size, embedding_size):
        super(Model, self).__init__()
        
        self.target = torch.randn((vocab_size, embedding_size), requires_grad = True)
        self.context = torch.randn((vocab_size, embedding_size), requires_grad = True)
    
    
    def forward(self, target_index, context_index):
        
        multiplication = torch.matmul(self.target[target_index], self.context.T)
        softmax = nn.functional.log_softmax(multiplication, dim=1)
        
        return softmax
        

model = Model(vocab_size, 150).cuda()      
```

```python
alpha = 0.003

optimizer = torch.optim.Adam([model.target, model.context], lr = alpha)
lossfn = nn.NLLLoss()
```

```python
epochs = 10
cost = []

for iteration in range(epochs):
    
    epoch_loss = 0
    for (center, context) in ls:
        
        center = torch.tensor([word2int[text[center]]])
        loss_target = torch.tensor([word2int[text[context]]])
        
        similarity_score = model.forward(center, loss_target).reshape(1,-1)
        loss = lossfn(similarity_score.cuda(), loss_target.cuda())
                
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        cost.append(loss.item())
        epoch_loss += loss.item()
        
    
    embeddings = model.target + model.context
    ls_words = ['french', 'revolution', 'revolutionary', 'government']
    words[0,:] = embeddings[word2int['french'],:]
    words[1,:] = embeddings[word2int['revolution'],:]
    words[2,:] = embeddings[word2int['revolutionary'],:]
    words[3,:] = embeddings[word2int['government'],:]

    pca = PCA(n_components = 2)
    pc_analysis = pca.fit_transform(words.detach().numpy())
    
    for i in range(len(pc_analysis)):
        plt.scatter(pc_analysis[i][0], pc_analysis[i][1])
        plt.annotate( ls_words[i], (pc_analysis[i][0], pc_analysis[i][1]))
    
    plt.show()
    
    print(f'total loss is {epoch_loss}')
```
